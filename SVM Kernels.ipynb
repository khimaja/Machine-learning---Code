```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from matplotlib import pyplot as plt
```

```python
# Generate two-dimensional classification dataset with 16 samples and two classes
X, y = make_blobs(n_samples=16, n_features=2, centers=2, random_state=42)

# Define different SVM kernels to plot
kernels = ["linear", "rbf", "poly"]

# Loop through each kernel and plot decision boundary and data points
for kernel in kernels:
    # Train SVM classifier with selected kernel
    clf = SVC(kernel=kernel)
    clf.fit(X, y)

    # Predict labels for all points on a grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

     # Plot decision boundary
    plt.contourf(xx, yy, Z, cmap="coolwarm", alpha=0.4)
    plt.plot(xx[Z == 0], yy[Z == 0], "o", color="pink", markersize=8, label="Class 0")
    plt.plot(xx[Z == 1], yy[Z == 1], "o", color="black", markersize=8, label="Class 1")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"SVM with {kernel} kernel")
    plt.legend()
    plt.show()
```
