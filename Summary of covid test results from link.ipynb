```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Madhya_Pradesh')
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.text, 'html.parser', from_encoding='utf-8')
```

finding all text in p, table and anchor tags:

```python
all_text = ''
for paragraph in soup.find_all('p'):
    all_text += paragraph.text
```

```python
for paragraph in soup.find_all('a'):
    all_text += paragraph.text
```

```python
for paragraph in soup.find_all('td'):
    all_text += paragraph.text
```

```python
all_text
```

```python
f = open('all_text.txt', 'w', encoding='utf-8')
f.write(all_text)
f.close()
```

<hr>

finding all links:

```python
all_links = ''
for links in soup.find_all('a', href=True):
    all_links += links['href']+'\n'
```

```python
all_links
```

```python
f = open('all_links.txt', 'w', encoding='utf-8')
f.write(str(all_links))
f.close()
```

finding summary

```python
for table_elem in soup.find_all('table', {'class': 'wikitable'}):
  summary = table_elem.text
```

```python
f = open('summary.txt', 'w', encoding='utf-8')
f.write(str(summary))
f.close()
```
